# Demoing house csv file

```python
house_df1 = pd.read_csv('house_sales.csv', header=0)
print(house_df1)
print(house_df1.isnull().sum())
```

```python
house_cols_lst = house_df1.columns.values.tolist()
print(house_cols_lst)
print(len(house_cols_lst))
```

```python
house_df1.loc[:, house_cols_lst].corr() # generate correlation matrix
```

```python
house_df2 = np.array(house_df1)
fig, axs = plt.subplots(4, 3, figsize=(20 , 12), sharey=False, sharex=False) # set figure fram
#house_df1['price'] = house_df1['price'].fillna(0)
axs[1, 1].boxplot(house_df1['price'].dropna()) # subplot 1 for full dataset
axs[1, 1].set_title('Boxplt for price')

axs[0, 1].boxplot(house_df1['bedrooms'].dropna()) # subplot 2 for outlier dataset
axs[0, 1].set_title('Boxplot for bedrooms')

axs[0, 2].boxplot(house_df1['bathrooms'].dropna()) # subplot 3 for outlier dataset
axs[0, 2].set_title('Boxplot for bathrooms')

axs[1, 0].boxplot(house_df1['sqft_living'].dropna()) # subplot 3 for outlier dataset
axs[1, 0].set_title('Boxplot for sqft_living')

axs[0, 0].boxplot(house_df1['sqft_lot'].dropna()) # subplot 3 for outlier dataset
axs[0, 0].set_title('Boxplot for sqft_lot')

axs[1, 2].boxplot(house_df1['floors'].dropna()) # subplot 3 for outlier dataset
axs[1, 2].set_title('Boxplot for floors')

axs[2, 0].boxplot(house_df1['condition'].dropna()) # subplot 3 for outlier dataset
axs[2, 0].set_title('Boxplot for condition')

axs[2, 1].boxplot(house_df1['grade'].dropna()) # subplot 3 for outlier dataset
axs[2, 1].set_title('Boxplot for grade')

axs[2, 2].boxplot(house_df1['sqft_above'].dropna()) # subplot 3 for outlier dataset
axs[2, 2].set_title('Boxplot for sqft_above')

axs[3, 0].boxplot(house_df1['sqft_basement'].dropna()) # subplot 3 for outlier dataset
axs[3, 0].set_title('Boxplot for sqft_basement')

axs[3, 1].boxplot(house_df1['sqft_living15'].dropna()) # subplot 3 for outlier dataset
axs[3, 1].set_title('Boxplot for sqft_living15')

axs[3, 2].boxplot(house_df1['sqft_lot15'].dropna()) # subplot 3 for outlier dataset
axs[3, 2].set_title('Boxplot for sqft_lot15')
plt.show()
```

```python
house_df3 = house_df1.dropna()
h_slot_q3, h_slot_q1 = np.percentile(house_df3['sqft_lot'], [75, 25]) # calculate quartiles 1 & 3
h_slot_iqr = h_slot_q3 - h_slot_q1 # calculate interquartile range
print('\nIQR: {}-{} = {}'.format(round(h_slot_q1, 4),round(h_slot_q3, 4),round(h_slot_iqr, 4))) # display IRQ

h_slot_iqr_out = h_slot_iqr * 1.5 # calculate outlier threshold
h_slot_otlr_low = h_slot_q1 - h_slot_iqr_out # calculate lower outlier limit
h_slot_otlr_high = h_slot_q3 + h_slot_iqr_out # calculate upper outlier limit
house_df3_sub1 = house_df3.loc[(house_df3['sqft_lot'] < h_slot_otlr_low) | (house_df3['sqft_lot'] > h_slot_otlr_high)] # use .loc method to search for records that are outliers; assign to new dataframe
house_df3_sub2 = house_df3.loc[(house_df3['sqft_lot'] >= h_slot_otlr_low) & (house_df3['sqft_lot'] <= h_slot_otlr_high)] # use .loc method to search for records that are outliers; assign to new dataframe
#print(house_df3_sub1) # display new dataframe

fig2, axs = plt.subplots(1, 3, sharey=True, figsize=(12 , 10)) # set figure fram
axs[0].boxplot(house_df3['sqft_lot'].dropna()) # subplot 1 for full dataset
axs[0].set_title('Sqft_lot Boxplot: Full Dataset')
axs[1].boxplot(house_df3_sub1['sqft_lot'].dropna()) # subplot 2 for outlier dataset
axs[1].set_title('Sqft_lot Boxplot: Outliers Subset')
axs[2].boxplot(house_df3_sub2['sqft_lot'].dropna()) # subplot 2 for outlier dataset
axs[2].set_title('Sqft_lot Boxplot: W/O Outliers Subset')
plt.show()
house_df3.describe()
print(house_df3['sqft_lot'].describe()) # descriptive stats for CRIM varibale
print(house_df3_sub1['sqft_lot'].describe()) # display descriptive stats of data subset
print(house_df3_sub2['sqft_lot'].describe()) # display descriptive stats of data subset

print('\nhouses mean age = {}'.format(round(house_df3['sqft_lot'].mean(), 4))) # average age for full dataset
print('houses median age = {}'.format(round(house_df3['sqft_lot'].median(), 4))) # average age for full dataset

print('\nhouses outliers mean age = {}'.format(round(house_df3_sub1['sqft_lot'].mean(), 4))) # average age for outliers
print('houses outliers median age = {}'.format(round(house_df3_sub1['sqft_lot'].median(), 4))) # average age for outliers
print('Sub1 Column count = {}'.format(len(house_df3_sub1.columns))) # alternative way to print only number of columns
print('Sub1 Row count = {}'.format(len(house_df3_sub1))) # alternative way to print only number of rows

print('\nhouses w/o outliers mean age = {}'.format(round(house_df3_sub2['sqft_lot'].mean(), 4))) # average age for outliers
print('houses w/o outliers median age = {}'.format(round(house_df3_sub2['sqft_lot'].median(), 4))) # average age for outliers
print('Sub2 Column count = {}'.format(len(house_df3_sub2.columns))) # alternative way to print only number of columns
print('Sub2 Row count = {}'.format(len(house_df3_sub2))) # alternative way to print only number of rows
```
